```python
#kategorisasikan jumlah retweet berdasarkan kategori, 0, 100, 500, 1000, di atas 1000 untuk masing-masing dataframe
#define fungsi kategori terlebih dahulu
def performance(total):
    if total < 100:
        return '-100'
    elif total > 100 and total < 500:
        return '100+'
    elif total >500 and total < 1000:
        return '500+'
    else:
        return '1000+'
```

```python
#kategorisasikan untuk performa retweet masing-masing dataframe
r1 = rejkt48time['retweets'].apply(performance).value_counts().plot(kind='bar',alpha=0.5,color='blue')
r2 = prejkt48time['retweets'].apply(performance).value_counts().plot(kind='bar',alpha=0.5)
r = plt.legend()
r.get_texts()[0].set_text('Post Reboost')
r.get_texts()[1].set_text('Pre Reboost')
r1.set_ylabel('Jumlah Kicauan')
r1.set_xlabel('Banyak Retweet')
```

```python
#kategorisasikan untuk performa like masing-masing dataframe
l1 = rejkt48time['likes'].apply(performance).value_counts().plot(kind='bar',alpha=0.5,color='yellow')
l2 = prejkt48time['likes'].apply(performance).value_counts().plot(kind='bar',alpha=0.5,color='red')
r = plt.legend()
r.get_texts()[0].set_text('Post Reboost')
r.get_texts()[1].set_text('Pre Reboost')
l1.set_ylabel('Jumlah Kicauan')
l2.set_xlabel('Banyak Likes')
```

```python
#kategorisasikan untuk performa reply masing-masing dataframe
rp1 = rejkt48time['replies'].apply(performance).value_counts().plot(kind='bar',alpha=0.5,color='black')
rp2 = prejkt48time['replies'].apply(performance).value_counts().plot(kind='bar',alpha=0.5,color='white')
r = plt.legend()
r.get_texts()[0].set_text('Post Reboost')
r.get_texts()[1].set_text('Pre Reboost')
rp1.set_ylabel('Jumlah Kicauan')
rp2.set_xlabel('Banyak replies')
```

```python
#Membandingkan rata-rata dua dataframe
m1 = rejkt48.mean().plot(legend=True)
m2 = prejkt48.mean().plot(legend=True)
r=plt.legend()
r.get_texts()[0].set_text('Post Reboost')
r.get_texts()[1].set_text('Pre Reboost')
m1.set_ylabel('Jumlah Kicauan')
m1.set_xlabel('Rata-Rata')
plt.savefig('tabel_rata2_RLRT')
```
